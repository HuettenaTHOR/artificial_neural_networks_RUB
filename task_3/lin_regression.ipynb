```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
size = 100
x = np.linspace(-3, 3, size)
eps = np.random.uniform(-0.5, 0.5, size)
```

```python
true_params = [5, 4]
Y = true_params[0] + true_params[1] * x + eps
```

```python
plt.plot(x, Y)
```

```python
N = size
cov_x_y = np.sum(x*Y) - (1/N)*(np.sum(x)*np.sum(Y))
var_x = (np.sum(x**2)-1/N*(np.sum(x)**2))
m = cov_x_y / var_x
print("m: ", m)
b = np.mean(Y) - m * np.mean(x)
print("b: ", b)
```

```python
bias = np.ones(size)
X = np.vstack((bias, x)).T
```

```python
theta_head = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)),X.T),Y)
```

```python
print(theta_head)
```

```python
plt.scatter(x, Y, c="blue")
y_estimate = theta_head[1]*x+theta_head[0]
plt.plot(x, y_estimate, c="red")
```

```python
SS_data = np.sum((Y-np.mean(Y))**2)
SS_reg = np.sum((y_estimate-np.mean(Y))**2)
R_two = SS_reg/SS_data
```
