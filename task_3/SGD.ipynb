```python
def l2_norm(true_params, pred_params):
    return np.linalg.norm(pred_params-true_params, 2)
```

```python
def generate_loss(n):
    N = n
    x = np.linspace(-2, 2, N)
    bias = np.ones(N)
    X = np.vstack((bias, x)).T
    eps = np.random.uniform(-0.5, 0.5, N)
    true_params = [5, 4]
    Y = true_params[0] + true_params[1]*x + eps
    epochs = 5
    lr = 0.1
    theta = np.zeros(2)
    for epoch in range(epochs):
        p = np.random.permutation(X.shape[0])
        _x = X[p]
        _y = Y[p]
        for i in range(x.shape[0]):
            theta = theta + lr*(_y[i]-np.dot(_x[i], theta))*_x[i]
    return theta


losses = []
l_x = []
for i in range(5, 500):
    l_x.append(i)
    losses.append(l2_norm([5, 4], generate_loss(i)))
    

plt.plot(l_x, losses)
        
```
