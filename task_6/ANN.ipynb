```python
import numpy as np
from scipy.special import expit
from matplotlib import pyplot as plt
```

```python
def approximation_function(x, v, bias, weight, N):
    sum = 0
    for i in range(N):
        sum += v[i]*expit(weight[i]*x+bias[i])
    return sum
```

```python
x = np.linspace(-2.5, 2.5, 100)
N = 1
b = np.array([0.])
v = np.array([1.])
w = [1, 10, 100, 1000]

fig, ax = plt.subplots()

for w_ in w:
    ax.plot(x, approximation_function(x, v, b, [w_], N), label=r"$w=%d$" %(w_))

ax.legend()
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
```

```python
w = np.array([1000])
b = np.array([-0.5])*w

fig, ax = plt.subplots()
ax.plot(x, approximation_function(x, v, [0], w, N), label = "f(x)")
ax.plot(x, approximation_function(x, v, b, w, N), label="f(x + c)")
```
